## Analyzing data from multiple files

```python
import glob
import matplotlib.pyplot as plt
```

```python
# glob statement is similar to a wild card

# Now we want to get all the filenames from the data directory
# then we want to loop (repeat the same operation on every file)
# that operation is that we want to plot the average inflammation rate per day
# the max and min inflamation rate per file
```

```python
filepattern = os.path.join("inflammation", "inflammation*.csv")
filenames = sorted(glob.glob(filepattern))
```

```python
# take the first three filenames
filenames = filenames [0:3]
for filename in filenames:
    print(filename)
    # Read data
    data = np.loadtxt(fname=filename, delimiter=",")
    
    # vizualise
    fig = plt.figure(figsize=(10.0, 3.0))
    
    # add axes
    # we are creating a grid where there are 1 row and three
    # columns, and we do this three times because
    # we want to investigate three things
    axes1 = fig.add_subplot(1,3,1)
    axes2 = fig.add_subplot(1,3,2)
    axes3 = fig.add_subplot(1,3,3)
    
    # add mean
    axes1.set_ylabel("average")
    meanval = np.mean(data, axis=0)
    axes1.plot(meanval)
    
    #add
    axes2.set_ylabel("max")
    maxval = np.max(data, axis=0)
    axes2.plot(maxval)
    
    axes3.set_ylabel("min")
    minval = np.min(data, axis=0)
    axes3.plot(minval)
    
    fig.tight_layout()
    plt.show()
```

## Creating functions

```python
def fahr2celcius(temp):
    """ transform fahrenheit temperature to celcius
    parameters:
        temp: numerical value
    """
    celcius = ((temp-32)*(5/9))
    return celcius
```

```python
fahr = 105
celcius = fahr2celcius(fahr)
print(celcius)
```

## Clean up visualization function

```python
def visualize(filename):
    """ Visualization of mean, max and min values in tabular data
    at file level
        Parameter:
            filename: str variable, path to file/tabular data object
    """
    # import data
    data = np.loadtxt(fname=filename, delimiter=",")
    
    # make canvas
    fig = plt.figure(figsize=(10.0,3.0))
    
    # create axes with 1 row, three columns, first/second/third figure
    axes1 = fig.add_subplot(1,3,1)
    axes2 = fig.add_subplot(1,3,2)
    axes3 = fig.add_subplot(1,3,3)
    
    # plot mean
    axes1.set_ylabel("Average")
    axes1.set_xlabel("Days")
    axes1.plot(np.mean(data, axis=0))
    
    # plot max
    axes2.set_ylabel("Max")
    axes2.set_xlabel("Days")
    axes2.plot(np.max(data, axis=0))
    
    # plot min
    axes3.set_ylabel("Min")
    axes3.set_xlabel("Days")
    axes3.plot(np.min(data, axis=0))
    
    # show
    fig.tight_layout()
    plt.show()
```

```python
filepattern = os.path.join("inflammation", "inflammation*.csv")
filenames = sorted(glob.glob(filepattern))
for filename in filenames:
    visualize(filename)
```
