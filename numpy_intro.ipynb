## Image translation

```python
# taking an image that looks like one thing
# and making it look like another
```

```python
path_to_image = os.path.join("..", "data", "img", "trex.png")
image = cv2.imread(path_to_image)
jimshow(image, "Original")
```

__Shifting image down and right__

__Translation matrix__

shift 25 pixels on x-axis, 50 pixels on y-axis

```python
# create an array in numpy
# Define a matrix M
# We want to shift the image 25 to the right
# and 50 pixels on the y-axis


M = np.float64([[1, 0, 25],
               [0, 1, 50]])
```

```python

width = image.shape[1]
height = image.shape[0]
```

```python
# Affine transformation, where everything keeps the same
# Anytime we want to perform a tranformation of an image
# we have to use cv2.

shifted = cv2.warpAffine(image, M, (width, height))
```

```python
# we call jimshow on the shifted image
jimshow(shifted, "Shifted down and right")
```

We can see the image is displaced. We have done this by using the translation matrix.
This is just an example of matrix modification, using MATHS!

__Shift up and left__

100 pixels left
100 pixels up

```python
# Here we want to use negative values to move the opposite way

M = np.float64([[1, 0, -100],
               [0, 1, -100]])
```

```python
shifted = cv2.warpAffine(image, M, (width, height))
```

```python
jimshow(shifted, "Shifted up and left")
```

__What does this do?__

```python

M = np.float64([[1, 0, 0],
               [0, 1, 100]])
shifted = cv2.warpAffine(image, M, (width, height))
```

```python
jimshow(shifted, "Shifted down")
```

__Writing a function for images__

```python
# Write a function that takes an image, a value for shifting
# the y and x axes and return a shifted image
# (image.shape shows the dimension of an image array
# so the 1 and 0 in the function below is asking
# for the first and second values of the image dimensions)


# the function takes three parameters:
# (1) the image,(2) the number of pixels on the x axis
# (3) the number of pixels on the y axis
# then it shifts along the width and height 
# and returns a shifted image

def translate(image, x, y): # define translation matrix
    M = np.float64([[1, 0, x], 
                   [0, 1, y]])
    
    shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0])) # perform translation on chosen image
            
    return shifted #return translated image
```

```python
shifted_image = translate(image, x=0, y=25)
```

```python
jimshow(shifted_image, "Shifted")
```
